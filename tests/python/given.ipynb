```python
import spot
from spot.jupyter import display_inline
spot.setup()
```

```python
aut1 = spot.translate('p | (!p & G(Fq & F!q & Fa))')
aut2 = spot.update_bounds_given(aut1, 'G(q)')
aut3 = spot.bounds_simplify(aut2)
display_inline(aut1, aut2, aut3)
```

```python
x = spot.translate('G(b -> Xd)'); x
```

```python
aut1 = spot.translate('a U b U c U d')
aut2 = spot.update_bounds_given(aut1, '!a')
aut3 = spot.update_bounds_given(aut2, x)
aut4 = spot.bounds_simplify(aut3)
display_inline(aut1, aut2, aut3, aut4, show='.h')
```

```python
common = spot.product(spot.translate("!a"), x) # Union of all knowledges
```

```python
spot.are_equivalent(spot.product(aut1, common), spot.product(aut4, common))
```

```python
aut1 = spot.translate('F(!p0 | X!p1)')
aut2 = spot.update_bounds_given(aut1, 'p0')
aut3 = spot.update_bounds_given(aut2, 'p1')
aut4 = spot.bounds_simplify(aut3)
display_inline(aut1, aut2, aut3, aut4)
```
